```python
import pandas as pd

# Load the main CSV and attribute info CSV
main_csv_path = "/Users/susanketsarkar/Desktop/Code/Meesho/data/train.csv"  # Replace with your main CSV path
attr_info_csv_path = "/Users/susanketsarkar/Desktop/Code/Meesho/data/attr_info.csv"  # Replace with your attr_info CSV path

# Function to generate separate CSV files for each category
def generate_category_csvs(main_csv_path, attr_info_csv_path):
    # Load data
    main_df = pd.read_csv(main_csv_path)
    attr_info_df = pd.read_csv(attr_info_csv_path)

    # Iterate through each row in the attribute info DataFrame
    for _, row in attr_info_df.iterrows():
        category = row['Category']
        no_of_attributes = row['No_of_attribute']
        attributes = row['Attribute_list'].replace(",", "").replace("[", "").replace("]", "").replace("'", "").replace("\"", "").replace("\n", "").split(" ")  # Convert string representation of list to actual list
        print(attributes)

        # Filter main DataFrame for the current category
        category_df = main_df[main_df['Category'] == category].copy()

        # Create new attribute columns based on the specified attributes
        for i in range(no_of_attributes):
            if i < len(attributes):
                # Create a new attribute column with the name from the attributes list
                category_df[attributes[i]] = category_df[f'attr_{i + 1}']
            else:
                # Fill with None if there are fewer attributes than specified
                category_df[attributes[i]] = None  

        # Drop the original attribute columns if necessary
        for i in range(1, 11):  # Assuming max 10 attributes based on your main CSV structure
            if f'attr_{i}' in category_df.columns:
                category_df.drop(columns=[f'attr_{i}'], inplace=True)

        # Save the filtered DataFrame to a new CSV file for the current category
        output_csv_path = f"../data/{category.replace(' ', '_')}_data.csv"  # Replace spaces in category names with underscores
        category_df.to_csv(output_csv_path, index=False)
        print(f"Generated CSV for category: {category}")

# Call the function
generate_category_csvs(main_csv_path, attr_info_csv_path)
```
